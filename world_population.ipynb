```python
import pandas as pd
import csv
import numpy as np

import psycopg2
from sqlalchemy import create_engine
```

## Data Extraction & Inspection - PostgreSQL Database

```python
# Establishing a connection to the database using create_engine.
engine = create_engine('postgresql://postgres:Thea1204%40@localhost:5432/dataengineering')
```

```python
# Defining the SQL query that will be used to extract the data from the database.
query = "SELECT * FROM world_population"
```

```python
# Fetch the data required from the database and store them in a pandas dataframe.
df = pd.read_sql(query, engine, index_col = 'Continent')
df.sort_values(by = ['Continent', 'Country/Territory', 'Rank'], ascending = True, inplace = True)
```

```python
# Closing the database
engine.dispose()
```
